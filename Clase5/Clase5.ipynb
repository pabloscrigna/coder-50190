#### Sentencia while

Vamos a comenzar con la sentencia iterativa m√°s b√°sica __while__ (mientras).
Se basa en repetir un bloque de c√≥digo a partir de evaluar una condici√≥n l√≥gica, siempre que esta sea True (al igual que la sentencia if).



image.png

Como programadores debemos decidir el momento en que la condici√≥n cambie a False para hacer que el while finalice su ejecuci√≥n y as√≠ salir de la iteraci√≥n, de lo contrario estar√≠amos frente a un bucle infinito.


Flujo de Ejecucion

1. Eval√∫a la condici√≥n, devolviendo True o False.


2. Si la condici√≥n es False, se sale de la sentencia While y contin√∫a la ejecuci√≥n con la siguiente sentencia.


3. Si la condici√≥n es True, ejecuta cada una de las sentencias en el bloque de c√≥digo y regresa al paso 1.

##### Continue

La instrucci√≥n continue da la opci√≥n de omitir la parte de un bucle en la que se activa una condici√≥n externa, pero continuar para completar el resto del bucle. Es decir, la iteraci√≥n actual del bucle se interrumpir√°, pero el programa volver√° a la parte superior del bucle.

Debe poner la instrucci√≥n continue dentro del bloque de c√≥digo bajo la instrucci√≥n de su bucle, generalmente despu√©s de una sentencia if condicional.


```python
# mejorar imprimiendo impares y el mensaje "numero par" si es par"
i = 0
while i < 6:
  i += 1
  if i == 3:
    continue
  print(i)
```

##### pass

Cuando se activa una condici√≥n externa, la instrucci√≥n pass permite manejar la condici√≥n sin que el bucle se vea afectado de ninguna manera; todo el c√≥digo continuar√° ley√©ndose a menos que se produzca la instrucci√≥n break u otra instrucci√≥n.

Debes poner la instrucci√≥n pass dentro del bloque de c√≥digo bajo la instrucci√≥n de su bucle, generalmente despu√©s de una sentencia if condicional.


```python
#Ejemplo 1
n = 0
while n < 10:
    n += 1
    if n == 2:
        pass
    print('n vale' , n)
```

```python
#Ejemplo 2
n = 0
while n < 10:
    n += 1
    if n == 2:
        pass
        print('n vale' , n)
```
